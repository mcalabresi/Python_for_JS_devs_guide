## Map, Filter and Reduce

map and filter are built-in functions while reduce comes from a library called functools

In the code below we make use of lambda functions as you may see. These are equivalent to anonymous functions in js

Notice how easy is to use range instead of Array.from , on the other side using map is a little more complicated in Python as you need to cast it to a list

```js
const numbers = Array.from({ length: 10 }, (v, i) => i);
console.log(numbers)
const doubleNumbers = numbers.map(n => 2*n)
console.log(doubleNumbers)

const oddNumbers = numbers.filter(n => n%2 !== 0)

const sumOfOddValues = oddNumbers.reduce((curr,acc)=> curr+acc,0)
```

Another thing to notice is that in Python land you will use == or != and not === or !== as we are used to in JS


````python
from functools import reduce

numbers = list(range(0,10))
print(numbers)

double_numbers = list(map(lambda x: 2*x, numbers))
print(double_numbers)

odd_numbers = list(filter(lambda x: x%2 != 0,numbers))
print(odd_numbers)

odd_numbers_sum = reduce(lambda curr,acc : curr+acc,odd_numbers,0)
print(odd_numbers_sum)



````

### List comprehensions
What if we could put map and filter capabilities in the same place and use a nicer syntax? That's what list comprehensions are for!!! Let's use list comprehensions to write the above examples. The important thing is to specify at the beginning what kind of value you want to be mapped and then for which values you want to loop for and in the end we pass an if clause to filter eventual unwanted values.

Once you learn list comprehensions please notice that we can use the same mechanism also for dicts ( dict comprehensions )

````python
numbers = list(range(0,10))
print(numbers)
# map with list comprehension
double_numbers = [2*x for x in numbers]
print(double_numbers)

# filter with list comprehension
odd_numbers = [x for x in numbers if x%2 !=0]
print(odd_numbers)

# map and filter together in the same list comprehension !
double_odd_numbers =  [2*x for x in numbers if x%2 != 0]
print (double_odd_numbers)
````

### generators

Maybe you heard about generators in JavaScript ( the star functions ), probably not but they are very useful. In a nutshell a generator is a function that returns an iterator. You don't return any value, but you yield it.
I'm not going to explain you what it is a generator, you can find more info here
=> https://developer.mozilla.org/en-US/docs/Web/JavaScript/Guide/Iterators_and_generators

Ok so python has got generators too, let's see the syntax and translate a JS generator in Python

JavaScript =>

```js
function* makeRangeIterator(start = 0, end = 10000, step = 1) {
  for (let i = start; i < end; i += step) {
    yield i;
  }
}

const generator = makeRangeIterator(1,10,2)

console.log(generator.next().value) // 1
console.log(generator.next().value) // 3
console.log(generator.next().value) // 5
```
Python =>


````python
def make_range_iterator(start=0, end=10000, step=1):
  for i in range(start,end,step):
    yield i

generator = make_range_iterator(start=1, end=10, step=2)

print(next(generator))
print(next(generator))
print(next(generator))


````

We can apply comprehensions to lists, dicts , sets and also generators. You should consider comprehensions as a crazy hybrid between map and filter functions. Some things to consider :
* the enclosing brackets define the type of data you want : square brackets is for lists, curly brackets for sets or dicts, round brackets for generator
* then you define the data that you want : one element normally, except two elements (key and value) for dicts
* the for cycle, specifying where to pick the input data from
* the if condition in case you want to filter out some data (optional)

Here some examples


````python
# list comprehension
res = [num**2 for num in range(1, 6)]
print(res)

# dict comprehension  (notice the element is expressed in the format key:value)
res = {num: num**3 for num in range(1, 6)}
print(res)

# set comprehension
a = [1, 2, 2, 3, 4, 4, 5, 6, 6, 7]
res = {num for num in a if num % 2 == 0}
print(res)

# generator comprehension
res = (num for num in range(10) if num % 2 == 0)
print(next(res))
print(next(res))
print(next(res))
# we put the other values into a list
print(list(res))
````

## OOP in Python

Object Oriented Programming in Python works as in Java, not like in JavaScript!

Quick recap about OOP in JS => In JavaScript OOP is prototypal and there are no "blueprints", just object instances. Keywords like class, constructor, extends are just syntactic sugar to cover quite a convoluted way of achieving OOP (also they were introduced in ES6, before we had only constructor functions). For further reading if you are not familiar with this check this article out => https://dev.to/jps27cse/understanding-javascript-prototypes-a-key-to-mastering-oop-1j1b

let's stick to some ES6 syntax for clarity and let's go on with our comparison in syntax between Python and JS

JS=>
```js
// creation of a Vehicle object

class Vehicle {
  constructor(type){
    this.type = type;
  }
  start(){
    console.log(`${this.type} vehicle starts`);
  }
  stop(){
    console.log(`${this.type} vehicle stops`);
  }
  // this is a special method that will be used when we want to cast this object to a string
  toString(){
    return `I am a ${this.type} vehicle `;
  }
}

const bicycle = new Vehicle('bicycle');
bicycle.start();
bicycle.stop();
// here we are casting bicycle to a string by prepending an empty string
console.log(''+bicycle);

// Car object now inherits from Vehicle object

class Car extends Vehicle {
  constructor(type,numberOfWheels){
    // important we have to invoke the constructor of the parent object
    super(type);
    this.numberOfWheels = numberOfWheels;
  }
  // here we are overloading the original toString method
  toString(){
    return super.toString() + ` and I have ${numberOfWheels} wheels`;
  }
}

const ferrari = new Car('car', 4);
// using the parent methods
ferrari.start();
ferrari.stop();
// now we use the overloaded method
console.log(''+ferrari);

```
Let's implement the same code in Python =>

````python
class Vehicle:
  # constructor method is __init__ . Remember that the first argument is always self (equivalent to this in JS)
  def __init__(self,type):
    self.type = type

  def start(self):
    print(f"{self.type} vehicle starts")

  def stop(self):
    print(f"{self.type} vehicle stops")
  # method to convert the object to a string
  def __str__(self):
    return f"I am a {self.type} vehicle"

bicycle = Vehicle('bicycle')
bicycle.start()
bicycle.stop()
print(str(bicycle))

class Car(Vehicle):
  def __init__(self,type,number_of_wheels):
    super().__init__(type)
    self.number_of_wheels = number_of_wheels

  def __str__(self):
    return super().__str__() + f" and I have {self.number_of_wheels} wheels"

ferrari = Car('car',4)
ferrari.start()
ferrari.stop()
print(str(ferrari))

````

## OOP Getters and setters

In JavaScript there is a nice way to create getters and setters for attributes of an object. Look at the example below
```js
let user = {
  name: "John",
  surname: "Smith",

  // after keywords get or set you should pass a function (method)
  get fullName() {
    return `${this.name} ${this.surname}`;
  },

  set fullName(value) {
    [this.name, this.surname] = value.split(" ");
  }
};

// The setter is executed with the given value. Under the hood we are going to process this value
user.fullName = "Alice Cooper";

console.log(user.fullName) // notice we treat fullName as a new attribute

console.log(user.name); // Alice
console.log(user.surname); // Cooper

```

Getters and setters are there for derived attributes.

Let's see how we can achieve the same in python ( in a pythonic way).
The best way is by using a decorator called @property

Python =>


````python
class user:
  def __init__(self,name,surname):
    self.name = name
    self.surname = surname
  # with this decorator we define the getter => in Python we are just saying
  # watch out, this is a new attribute of the class
  @property
  def full_name(self):
    return f"{self.name} {self.surname}"

  # here we address the previously defined attribute and we specify the setter
  # as in JS we should provide functions for getter and setter
  @full_name.setter
  def full_name(self,value):
    [self.name,self.surname] = value.split(" ")

# class in Python is just a blueprint , so we should instantiate and initialize
user = user("John","Smith")

# using the setter
user.full_name = "Alice Cooper"

# using the getter
print(user.full_name)

# the setter did the update job for us
print(user.name)
print(user.surname)
````

## Decorators

A decorator in Python is a function (or a class) that wraps another function (input function) to enhance its behaviour without modifying the input function. It is part of so-called metaprogramming and it's a very useful concept that is widely used in many languages.

The most modern syntax is not (yet) part of vanilla JavaScript but it's used extensively in TypeScript and uses the '@' symbol. Read more here => https://medium.com/@artemkhrenov/the-decorator-pattern-in-modern-javascript-adding-functionality-without-breaking-code-b43d9c237047

Notice that in TypeScript decorators are limited to classes, methods and attributes of an object


````python
# a very simple example of a decorator function

def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.\n")
    # we take as input a function and we return as output a wrapped function
    return wrapper

@my_decorator
def say_hello():
    print("Hello! I am the output of the original function")

say_hello()

# a very simple example of a decorator class

class deco_class:
  # we need to pass the input function as an attribute, we need to call it later
    def __init__(self, func):
        self.func = func
  # each time we call the decorated input function we execute this code
    def __call__(self):
        print("Hello I am a decorator class")
        # here we execute the input function
        self.func()
        print("Bye\n")

@deco_class
def say_hi():
    print("Hi, I am the output of the original function")

say_hi()
````

## Modules and packages

To create a package we have to create a directory and inside we have to place an empty file named "__init__.py".  from <module> import <function_name>

 from <package>.<subpackage> import <module>

 from <package>.<subpackage>.<module> import <function>

 There is quite a big freedom. You could also import all the functions from a module like so

 where * is the wildcard to import everything. YOU SHOULD NOT DO IT because you can mask some other definitions in standard library or precedingly loaded modules. Also memory-wise it is not a great choice.

## Walrus operator

This syntax is not present in JavaScript but it's quite interesting and Pythonic.

The Walrus operator, `:=`, allows you to assign a value to a variable within an expression. This can be useful for reducing code redundancy and making your code more concise, especially in `if` and `while` statements where you need to compute a value and then test it.

### Example:

Without the walrus operator:
```python
foods = []
while True:
    food = input("What food do you like? (type 'quit' to exit): ")
    if food == 'quit':
        break
    foods.append(food)
print(f"Your favorite foods are: {foods}")
```

With the walrus operator:
```python
foods = []
while (food := input("What food do you like? (type 'quit' to exit): ")) != 'quit':
    foods.append(food)
print(f"Your favorite foods are: {foods}")
```

````python
# Example of the Walrus operator in a while loop
foods = []
while (food := input("What food do you like? (type 'quit' to exit): ")) != 'quit':
    foods.append(food)
print(f"Your favorite foods are: {foods}")

# Example of the Walrus operator in an if statement
if (n := len(foods)) > 0:
    print(f"You entered {n} foods.")
else:
    print("You didn't enter any foods.")
````

## Writing and reading files in Python

If you worked in Node you know you can read or write on the file system using the fs library. Here a quick guide => https://www.w3schools.com/nodejs/nodejs_filesystem.asp

In a nutshell you want to know
* where the file is
* which kind of permissions you need to have to access it ( read-only, update, append, etc)
* the encoding of the file
* if you want to access synchronously or asynchronously

IMPORTANT
In Node you specify the path to the file, In Python you should first change work directory to the folder where the file you want to read or write, only then you can access the file

In Python, file operations are generally performed using the built-in `open()` function. This function returns a file object, which has methods for reading and writing data.

**Basic File Operations:**

*   **Opening a file:** `open(filename, mode)`
    *   `filename`: The path to the file.
    *   `mode`: Specifies how the file will be used. Common modes include:
        *   `'r'` (read): Opens for reading (default). Raises `FileNotFoundError` if the file doesn't exist.
        *   `'w'` (write): Opens for writing. Creates a new file if it doesn't exist, or truncates (empties) the file if it does. **Be careful as it overwrites existing content.**
        *   `'a'` (append): Opens for appending. Creates a new file if it doesn't exist. Writes new data to the end of the file.
        *   `'x'` (create): Creates a new file and opens it for writing. Raises `FileExistsError` if the file already exists.
        *   `'t'` (text): Default mode for text files.
        *   `'b'` (binary): For binary files (e.g., images, executables).

*   **Reading:**
    *   `file_object.read()`: Reads the entire content of the file.
    *   `file_object.readline()`: Reads a single line from the file.
    *   `file_object.readlines()`: Reads all lines into a list of strings.

*   **Writing:**
    *   `file_object.write(string)`: Writes the given string to the file.

*   **Closing a file:** `file_object.close()`
    *   It's crucial to close files to free up system resources and ensure all changes are saved. Failing to close files can lead to data corruption or resource leaks.

### The `with` Statement for File Handling

The `with` statement is the recommended way to handle file operations in Python because it ensures that the file is automatically closed, even if errors occur. This is known as a "context manager."

**Syntax:**
```python
with open(filename, mode) as file_object:
    # Perform read/write operations here
    pass
# The file is automatically closed when exiting the 'with' block
```

Using `with` makes your code safer and cleaner, as you don't have to explicitly call `file_object.close()`.

````python
# --- Writing to a file ---
file_content = "Hello, Python!\nThis is a test file.\nLearning to write files is fun."

# Using 'w' mode to write (overwrites existing file or creates a new one)
with open('my_example_file.txt', 'w') as f:
    f.write(file_content)
print("Content successfully written to my_example_file.txt")

# --- Appending to a file ---
append_content = "\nThis line was appended."

# Using 'a' mode to append
with open('my_example_file.txt', 'a') as f:
    f.write(append_content)
print("Content successfully appended to my_example_file.txt")

# --- Reading from a file ---
# Using 'r' mode to read
with open('my_example_file.txt', 'r') as f:
    read_data = f.read()
    print("\n--- Content of my_example_file.txt ---")
    print(read_data)

# --- Reading line by line ---
print("\n--- Reading line by line ---")
with open('my_example_file.txt', 'r') as f:
    for line in f:
        print(f"Line: {line.strip()}") # .strip() removes leading/trailing whitespace, including newline characters
````

### `random`
This module implements pseudo-random number generators for various distributions. It's used for tasks like shuffling lists or picking random elements.

````python
import random

# Get a random integer between 1 and 10
print(random.randint(1, 10))

# Pick a random element from a list
items = ['apple', 'banana', 'cherry']
print(random.choice(items))
````

### `re` (Regular Expression)
This module provides regular expression operations. It's essential for pattern matching and string manipulation.

````python
import re

text = "The quick brown fox."
# Search for 'quick' in the text
match = re.search(r'quick', text)
print(match.group() if match else "Not found")

# Replace 'fox' with 'dog'
new_text = re.sub(r'fox', 'dog', text)
print(new_text)
````

### `copy`
This module provides generic shallow and deep copy operations. It's used to create independent duplicates of mutable objects.

````python
import copy

original_list = [1, [2, 3], 4]

# Shallow copy: changes to inner lists affect both
shallow_copied_list = copy.copy(original_list)
shallow_copied_list[1][0] = 99
print(f"Original after shallow copy modification: {original_list}")

# Deep copy: fully independent
deep_copied_list = copy.deepcopy(original_list)
deep_copied_list[1][0] = 100
print(f"Original after deep copy modification: {original_list}")
````

### `functools`
This module provides higher-order functions and operations on callable objects. It's useful for function decorators and partial application.

````python
import functools

# Use functools.partial to create a new function with some arguments pre-filled
def multiply(a, b):
    return a * b

double = functools.partial(multiply, 2)
print(f"Double of 5: {double(5)}")
````

### `numpy` (Numerical Python)
This library adds support for large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays. It's fundamental for scientific computing in Python.

````python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print(f"NumPy array: {arr}")

# Perform element-wise operations
print(f"Array multiplied by 2: {arr * 2}")
print(f"Sum of array elements: {np.sum(arr)}")
````
